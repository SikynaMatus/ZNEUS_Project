```python
def get_upper_lower_iqr(df: pd.DataFrame, col: str):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return (lower_bound, upper_bound)
```

```python
#statistical power
def cohen_d(x1, x2):
    nx1 = len(x1)
    nx2 = len(x2)
    s = np.sqrt(((nx1-1) * np.std(x1, ddof=1)**2 + (nx2-1) * np.std(x2, ddof=1)**2) / (nx1 + nx2 - 2))
    return (np.abs(np.mean(x1) - np.mean(x2))) / s
def calculate_statistical_power(i1, i2):
    x = np.asarray(i1.dropna())
    y = np.asarray(i2.dropna())

    cohen = cohen_d(x, y)

    alpha = 0.05
    power = sm_power.tt_ind_solve_power(cohen, len(x), alpha, None, 1)

    print(f"Power = {power:.3f}\n")

    return power
```

```python
#Class for migrating to PyTorch dataset
class ToTorchDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
```

```python
#Creating tensors from dataset values
X_train_tensor = torch.tensor(X_train_testing.values, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test_testing.values, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train_testing.values, dtype=torch.float32).view(-1,1)
y_test_tensor = torch.tensor(y_test_testing.values, dtype=torch.float32).view(-1,1)

#Migrating to PyTorch
train_dataset = ToTorchDataset(X_train_tensor, y_train_tensor)
test_dataset = ToTorchDataset(X_test_tensor, y_test_tensor)

#initializing loaders as function, just to use config
def loading_data(train_dataset, test_dataset, cfg):
  train_loader = DataLoader(train_dataset, cfg.batch_size, shuffle=True, num_workers=cfg.num_workers)
  test_loader = DataLoader(test_dataset, cfg.batch_size, shuffle=False, num_workers=cfg.num_workers)
  return train_loader, test_loader
```

```python
#From NNPI-01_Practical_Intro

class Statistics:
  def __init__(self):
    self.values = dict()

  def step(self, key, value):
    sum, count = 0.0, 0.0
    if key in self.values:
      sum, count = self.values[key]
    sum += value
    count += 1
    self.values[key] = (sum, count)

  def get(self):
    result = dict()
    for k, (sum, count) in self.values.items():
      result[k] = float(sum / count)
    return result
```

```python
#From NNPI-01_Practical_Intro

class BinaryClassifier(nn.Module):
  def __init__(self, n_in, n_hidden, n_out, dropout=0.5, flag = 0):
    super().__init__()

    if flag==0:
      self.main = nn.Sequential(
        nn.Linear(n_in, n_hidden), #in this case, we aren't working with pictures
        nn.ReLU(),
        nn.Linear(n_hidden, n_hidden),
        nn.ReLU(),
        nn.Linear(n_hidden, n_out)
    )
    elif flag==1:
      self.main = nn.Sequential(
        nn.Linear(n_in, n_hidden),
        nn.BatchNorm1d(n_hidden),
        nn.ReLU(),
        nn.Linear(n_hidden, n_hidden),
        nn.ReLU(),
        nn.Linear(n_hidden, n_out)
    )
    elif flag==2:
      self.main = nn.Sequential(
        nn.Linear(n_in, n_hidden),
        nn.ReLU(),
        nn.Linear(n_hidden, int(n_hidden/2)),
        nn.ReLU(),
        nn.Linear(int(n_hidden/2), n_hidden),
        nn.ReLU(),
        nn.Linear(n_hidden, n_out)
    )
    elif flag==3:
      self.main = nn.Sequential(
        nn.Linear(n_in, n_hidden),
        nn.ReLU(),
        nn.Dropout(dropout),
        nn.Linear(n_hidden, n_hidden),
        nn.ReLU(),
        nn.Linear(n_hidden, n_out)
    )

  def forward(self, x):
    return self.main(x)
```

```python
#From NNPI-01_Practical_Intro

class Trainer:
  def __init__(self, model, cfg):

    self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    self.model = model.to(self.device)
    self.cfg = cfg

    self.opt = torch.optim.Adam(self.model.parameters(), lr=self.cfg.learning_rate) #I've heard it's usually better than SGD

    imbalance = torch.tensor(self.cfg.n_neg/self.cfg.n_pos, dtype=torch.float32).to(self.device) #this is supposed to fix imbalance between large difference between mach and no match
    self.loss = nn.BCEWithLogitsLoss(pos_weight= imbalance) #We are doing binary classification

    #either having imballance in "match" causes overfitting and 85% is acctually quite high, or I did something wrong.
    #self.loss = nn.BCEWithLogitsLoss()

    self.loader_train = None
    self.loader_test = None

    self.history = {"loss_train": [], "loss_test": [], "accuracy_train": [], "accuracy_test": []}

  def setup(self, train_dataset, test_dataset):
    self.loader_train, self.loader_test = loading_data(train_dataset, test_dataset, self.cfg)


  def fit(self):
    best_accuracy = 0
    best_loss = 100
    for epoch in range(self.cfg.max_epochs):

      stats = Statistics()
      with tqdm(
          self.loader_train,
          desc=f"Train: {epoch}"
      ) as progress:
        correct, total = 0, 0 #for additional metrics
        for x, y in progress:
          x = x.to(self.device)
          y = y.to(self.device)

          y_hat = self.model(x)
          l = self.loss(y_hat, y)

          self.opt.zero_grad()
          l.backward()
          self.opt.step()

          #Accuracy
          preds = (torch.sigmoid(y_hat) > 0.5).float()  #threshold
          correct += (preds == y).sum().item()
          total += y.size(0)

          stats.step("loss_train", l.item())
          progress.set_postfix(stats.get())

        acc_train = correct / total
        print(f"Epoch {epoch}: Train loss = {stats.get()['loss_train']:.4f}, Train accuracy = {acc_train:.4f}")

        self.history["loss_train"].append(stats.get()["loss_train"])
        self.history["accuracy_train"].append(acc_train)

      stats = Statistics()
      with torch.no_grad():

        with tqdm(
            self.loader_test,
          desc=f"Test: {epoch}"
        ) as progress:
          correct, total = 0, 0 #for additional metrics
          for x, y in progress:
            x = x.to(self.device)
            y = y.to(self.device)

            y_hat = self.model(x)
            l = self.loss(y_hat, y)

            #Accuracy
            preds = (torch.sigmoid(y_hat) > 0.5).float()  #threshold
            correct += (preds == y).sum().item()
            total += y.size(0)

            stats.step("loss_test", l.item())
            progress.set_postfix(stats.get())

          acc_test = correct / total
          if acc_test > best_accuracy or stats.get()["loss_test"] < best_loss: #to prevent "high accuracy" in the begining
            best_accuracy = acc_test
            best_loss = stats.get()["loss_test"]

          print(f"Epoch {epoch}: Test loss = {stats.get()['loss_test']:.4f}, Test accuracy = {acc_test:.4f}")

          self.history["loss_test"].append(stats.get()["loss_test"])
          self.history["accuracy_test"].append(acc_test)


    self.plot_history()
    return best_accuracy

  def plot_history(self):
    epochs = range(1, len(self.history["loss_train"]) + 1)
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(epochs, self.history["loss_train"], label="Train Loss")
    plt.plot(epochs, self.history["loss_test"], label="Test Loss")
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs, self.history["accuracy_train"], label="Train Acc")
    plt.plot(epochs, self.history["accuracy_test"], label="Test Acc")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.title("Accuracy over epochs")
    plt.legend()

    plt.tight_layout()
    plt.show()
```
